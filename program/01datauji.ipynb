```python
df_smiles = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/fitur_herb.csv')
df_smiles
```

```python
import ast
import re
def fix_fingerprint_format(fingerprint):
    fixed = re.sub(r"(?<=[\d])\s+(?=[\d])", ", ", fingerprint.strip())  # Tambahkan koma
    return fixed

df_smiles['pubchem_fingerprint'] = df_smiles['pubchem_fingerprint'].apply(fix_fingerprint_format)

# Mengubah string menjadi list Python
import ast
df_smiles['pubchem_fingerprint'] = df_smiles['pubchem_fingerprint'].apply(ast.literal_eval)

# Memisahkan fingerprint menjadi kolom-kolom individu
fingerprint_df = pd.DataFrame(
    df_smiles['pubchem_fingerprint'].tolist(),
    columns=[str(i) for i in range(len(df_smiles['pubchem_fingerprint'][0]))]
)

# Menggabungkan fingerprint dengan kolom PubChem_CID
result = pd.concat([df_smiles[['com_pubchem_id']], fingerprint_df], axis=1)
```

```python
result
```

```python
# Rename the 'com_pubchem_id' column to 'cid'
result = result.rename(columns={'com_pubchem_id': 'cid'})
result
result.to_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/fitur_herb_fix.csv', index=False)
```

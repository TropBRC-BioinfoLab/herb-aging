```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,auc,ConfusionMatrixDisplay
```

```python
train_aac = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/train_aac.csv')
train_dc = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/train_dc.csv')
train_aac_ro = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/train_aac_ro.csv')
train_dc_ro = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/train_dc_ro.csv')
train_aac_sm = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/train_aac_sm.csv')
train_dc_sm = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/train_dc_sm.csv')
test_aac = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/test_aac.csv')
test_dc = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/test_dc.csv')
```

# Imbalance

## AAC

```python
X_train_aac = train_aac.iloc[:,:-1]
y_train_aac = train_aac.iloc[:,-1]
X_test_aac = test_aac.iloc[:,:-1]
y_test_aac = test_aac.iloc[:,-1]
```

```python
X_train_aac
```

```python
y_train_aac
```

```python
X_test_aac
```

```python
y_test_aac
```

```python
rf = RandomForestClassifier()
n_estimators = [10,50,100]
max_depth = [10,50,100,None]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 3]

grid = {'n_estimators': n_estimators,
        'max_depth': max_depth,
        'min_samples_split': min_samples_split,
        'min_samples_leaf': min_samples_leaf}

rf_random = GridSearchCV(estimator = rf, param_grid = grid,
                         cv = 2, scoring ='f1', verbose=10)
```

```python
rf_random.fit(X_train_aac,y_train_aac)
print('Best Parameters: ', rf_random.best_params_, ' \n')
```

```python
model_aac = RandomForestClassifier(**rf_random.best_params_) # Use best parameters from grid search
model_aac.fit(X_train_aac,y_train_aac)
y_pred = model_aac.predict(X_test_aac)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,auc,ConfusionMatrixDisplay
print("Classification Report")
print(classification_report(y_pred,y_test_aac))
```

```python
print("Accuracy:",accuracy_score(y_pred,y_test_aac))
print("Precision :",precision_score(y_pred,y_test_aac))
print("Recall :",recall_score(y_pred,y_test_aac))
print("F-1 Score : ",f1_score(y_pred,y_test_aac))
print("AUROC:", roc_auc_score(y_pred,y_test_aac))
```

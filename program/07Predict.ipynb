```python
aac_interaction = predict_aac.sort_values(by='prob', ascending=False)
aac_interaction = aac_interaction.rename(columns={'cid': 'com_pubchem_id'})
aac_interaction
```

```python
aac_compound = aac_interaction.merge(compound, on = 'com_pubchem_id')
aac_compound = aac_compound[['protein','com_id','com_pubchem_id','prob']]
aac_compound
```

```python
aac_compound.to_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/aac_compound.csv', index = False)
```

```python
aac_plant = aac_compound.merge(plant, on = 'com_id')
aac_plant = aac_plant[['protein','com_id','com_pubchem_id','pla_id','prob']]
aac_plant
```

```python
aac_plant = aac_plant.merge(plant_name, on = 'pla_id')
aac_plant
```

```python
aac_plant = aac_plant[['protein','com_id','com_pubchem_id','pla_id','pla_name','pla_idr_name','prob']]
aac_plant
```

```python
aac_plant.to_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/aac_plant.csv', index = False)
```

PREDICT DC

```python
# Load model dari file
with open('/content/drive/MyDrive/Skripsi/Program/random_forest_best_dc.pkl', 'rb') as model_file:
    loaded_model_2 = pickle.load(model_file)

# Gunakan model untuk prediksi
y_pred = loaded_model_2.predict(herb_dc)
```

```python
y_probs_dc = loaded_model_2.predict_proba(herb_dc)[:, 1]
herb_dc['interaction'] = y_pred
herb_dc['prob'] = y_probs_dc

predict_dc = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/interaksi.csv')
predict_dc['interaction'] = y_pred
predict_dc['prob'] = y_probs_dc

predict_dc.to_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/predict_herbaldb_dc.csv', index = False)

dc_count = predict_dc[predict_dc.interaction == 1]['cid'].value_counts().head()

dc_interaction = predict_dc.sort_values(by='prob', ascending=False)
dc_interaction = dc_interaction.rename(columns={'cid': 'com_pubchem_id'})

dc_compound = dc_interaction.merge(compound, on = 'com_pubchem_id')
dc_compound = dc_compound[['protein','com_id','com_pubchem_id','prob']]
dc_compound.to_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/dc_compound.csv', index = False)

dc_plant = dc_compound.merge(plant, on = 'com_id')
dc_plant = dc_plant[['protein','com_id','com_pubchem_id','pla_id','prob']]
dc_plant = dc_plant.merge(plant_name, on = 'pla_id')
dc_plant = dc_plant[['protein','com_id','com_pubchem_id','pla_id','pla_name','pla_idr_name','prob']]
dc_plant.to_csv('/content/drive/MyDrive/Skripsi/Program/data/herb/dc_plant.csv', index = False)
```

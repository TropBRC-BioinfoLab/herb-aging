```python
# prompt: samakan dengan proses fp_dc sebelumnya

import numpy as np
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE

# Assuming fp_aac is already loaded as in the previous code

# Menampilkan baris yang memiliki missing values
rows_with_missing = fp_aac[fp_aac.isnull().any(axis=1)]
print("\nRows with missing values:")
print(rows_with_missing)

fp_aac = fp_aac.dropna()

# AAC Descriptor
fp_aac.interaction.value_counts().plot.bar()
print(fp_aac.interaction.value_counts(normalize = True))
print(fp_aac.interaction.value_counts())
x_aac = fp_aac.iloc[:,3:]
y_aac = fp_aac.iloc[:,2]

# Random Oversampling
ros = RandomOverSampler(random_state=42)
x_aac_res, y_aac_res = ros.fit_resample(x_aac, y_aac)

# Tampilkan hasil
print("Sebelum Oversampling:")
print(y_aac.value_counts())
print("\nSetelah Oversampling:")
print(pd.Series(y_aac_res).value_counts())

balanced_fp_aac = x_aac_res
balanced_fp_aac['interaction'] = y_aac_res
balanced_fp_aac.info()
balanced_fp_aac.interaction.value_counts().plot.bar()
balanced_fp_aac.to_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/balanced_fp_aac.csv', index = False)

# SMOTE
sm = SMOTE(random_state=42)
x_res2_aac,y_res2_aac = sm.fit_resample(x_aac,y_aac)

# Tampilkan hasil
print("Sebelum SMOTE:")
print(y_aac.value_counts())
print("\nSetelah SMOTE:")
print(pd.Series(y_res2_aac).value_counts())

balanced_fp_aac2 = x_res2_aac
balanced_fp_aac2['interaction'] = y_res2_aac
balanced_fp_aac2.info()
balanced_fp_aac2.interaction.value_counts().plot.bar()
balanced_fp_aac2.to_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/balanced_fp_aac2.csv', index = False)
```

```python
import pandas as pd
import ast
import re
```

```python
# Membaca file CSV
df = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/fitursenyawa.csv')
```

```python
# Perbaikan format pubchem_fingerprint: Menambahkan koma jika hilang
def fix_fingerprint_format(fingerprint):
    fixed = re.sub(r"(?<=[\d])\s+(?=[\d])", ", ", fingerprint.strip())  # Tambahkan koma
    return fixed

df['pubchem_fingerprint'] = df['pubchem_fingerprint'].apply(fix_fingerprint_format)

# Mengubah string menjadi list Python
import ast
df['pubchem_fingerprint'] = df['pubchem_fingerprint'].apply(ast.literal_eval)

# Memisahkan fingerprint menjadi kolom-kolom individu
fingerprint_df = pd.DataFrame(
    df['pubchem_fingerprint'].tolist(),
    columns=[str(i) for i in range(len(df['pubchem_fingerprint'][0]))]
)

# Menggabungkan fingerprint dengan kolom PubChem_CID
result = pd.concat([df[['PubChem_CID']], fingerprint_df], axis=1)
```

```python
result
```

```python
# Menyimpan hasil ke file CSV
result.to_csv('/content/drive/MyDrive/Skripsi/Program/data/fitursenyawa_fix.csv', index=False)
```

```python
df_interaction = pd.read_csv('/content/drive/MyDrive/Skripsi/Program/data/interactions.csv', index_col = [0])
df_interaction
```

```python
# Mapping uniprot_id yang ingin diubah
uniprot_mapping = {
    "Q96EB6": "SIR1",
    "Q8IXJ6": "SIR2",
    "Q9NTG7": "SIR3",
    "Q9Y6E7": "SIR4",
    "Q9NXA8": "SIR5",
    "P05019": "IGF1",
    "Q9UEF7": "KLOT",
    "Q8N726": "CDN2A",
    "Q14457": "BECN1",
    "P10415": "BCL2"
}

# Ganti nilai kategori berdasarkan mapping
df_interaction["UniprotID"] = df_interaction["UniprotID"].replace(uniprot_mapping)
df_interaction = df_interaction.rename(columns = {"UniprotID":"uniprot_id"})
df_interaction = df_interaction.rename(columns = {"PubChem_CID":"pubchem_id"})
df_interaction
```

```python
x_aac = fp_aac.iloc[:,3:]
x_aac
```

```python
y_aac = fp_aac.iloc[:,2]
y_aac
```

SMOTEENN

```python
from imblearn.combine import SMOTEENN
from collections import Counter
```

```python
# Inisialisasi SMOTEENN
smote_enn = SMOTEENN(random_state=42)

# Lakukan resampling pada dataset
x_aac_res, y_aac_res = smote_enn.fit_resample(x_aac, y_aac)

# Tampilkan distribusi kelas sebelum dan sesudah balancing
print("Sebelum SMOTEENN:", Counter(y_aac))
print("Setelah SMOTEENN:", Counter(y_aac_res))
```

```python
print("Sebelum Oversampling:")
print(y_aac.value_counts())
print("\nSetelah Oversampling:")
print(pd.Series(y_aac_res).value_counts())

balanced_fp_aac3 = x_aac_res
balanced_fp_aac3['interaction'] = y_aac_res
balanced_fp_aac3.info()
balanced_fp_aac3.interaction.value_counts().plot.bar()
balanced_fp_aac3.to_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/balanced_fp_aac3.csv', index = False)
```

```python
x_dc = fp_dc.iloc[:,3:]
y_dc = fp_dc.iloc[:,2]

# SMOTEENN
# Inisialisasi SMOTEENN
smote_enn = SMOTEENN(random_state=42)

# Lakukan resampling pada dataset
x_dc_res, y_dc_res = smote_enn.fit_resample(x_dc, y_dc)

# Tampilkan distribusi kelas sebelum dan sesudah balancing
print("Sebelum SMOTEENN:", Counter(y_dc))
print("Setelah SMOTEENN:", Counter(y_dc_res))
```

```python
print("Sebelum Oversampling:")
print(y_dc.value_counts())
print("\nSetelah Oversampling:")
print(pd.Series(y_dc_res).value_counts())

balanced_fp_dc3 = x_dc_res
balanced_fp_dc3['interaction'] = y_dc_res
balanced_fp_dc3.info()
balanced_fp_dc3.interaction.value_counts().plot.bar()
balanced_fp_dc3.to_csv('/content/drive/MyDrive/Skripsi/Program/data/dtp/balanced_fp_dc3.csv', index = False)
```

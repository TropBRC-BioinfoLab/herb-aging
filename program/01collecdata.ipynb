```python
def filter_data(df):
    df_new = df[['UniProt (SwissProt) Primary ID of Target Chain','UniProt (SwissProt) Entry Name of Target Chain','PubChem CID','Ligand SMILES']]
    df_new.rename(columns = {'UniProt (SwissProt) Primary ID of Target Chain':'UniprotID','UniProt (SwissProt) Entry Name of Target Chain':'Uniprot Name'},inplace = True)
    return (df_new)
```

```python
import requests
import pandas as pd
```

```python
# Fungsi untuk mendapatkan SMILES dari PubChem menggunakan CID
def get_smiles_from_cid(cid):
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/property/IsomericSMILES/JSON"
    response = requests.get(url)
    
    # Jika request berhasil
    if response.status_code == 200:
        data = response.json()
        if 'PropertyTable' in data and 'Properties' in data['PropertyTable']:
            return data['PropertyTable']['Properties'][0].get('IsomericSMILES', None)
    return None

# Membaca file CSV
df = pd.read_csv('data/PubChem.csv', delimiter=';')

# Loop melalui setiap CID dan mengambil SMILES
for index, row in df.iterrows():
    cid = row['PubChem_CID']
    smiles = get_smiles_from_cid(cid)
    df.at[index, 'Ligand SMILES'] = smiles

# Menampilkan DataFrame dengan SMILES yang telah diambil
print(df)

# Menyimpan hasil ke CSV
df.to_csv("output_with_smiles.csv", sep=';', index=False)
```
